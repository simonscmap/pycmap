## *Data Visualization*

The main focus of pycmap (and the broader Simons CMAP project) is to provide the user with a unifying API to extract subsets of datasets with different dimensions and different spatio-temporal resolutions. This will drastically reduce the time-consuming data collection and preparation processes, especially considering the massive size of the archived multi-decade global oceanic datasets generated by satellites or numerical models. 

<br />The simple and unified interface of the pycmap API enables the general public and scientists to dive into the vast, and often under underutilized, ocean datasets without being concerned about the underlying dataset file structure. Visualizing the retrieved data is often a good entry point to explore and study these datasets and pycmap offers a number of builtin methods for simple interactive data visualizations. The pycmap visualizetion engine can be set to either of these interactive data visualization libraries: [plotly](https://plot.ly/), or [bokeh](https://bokeh.pydata.org/en/latest/index.html). Plotly has a larger variation graphs while bokeh is faster to render the graph object. 

<br />The default visualization library is plotly; the code block below shows how to change the visualization engine (see [API](API.ipynb) page for more details).

<br />**Note:**
<br />Data visualization requires a valid [API key](API.ipynb) which can be obtained from [https://simonscmap.com/](https://simonscmap.com/). It is not necessary to set the API key every time because the API properties are stored locally after being called the first time.


```python
import pycmap
pycmap.API(vizEngine='bokeh')  # vizEngine options: 'plotly' ,'bokeh'
```
